```python
import pandas as pd
import plotly.express as px

# 1. Create a 12-month compliance dataset
compliance_data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Percent_Compliant': [65, 68, 72, 70, 75, 82, 80, 85, 88, 92, 91, 95],
    'Open_Audit_Findings': [15, 12, 10, 11, 8, 5, 6, 4, 3, 2, 2, 1]
}

df_trends = pd.DataFrame(compliance_data)

# 2. Create an interactive Line Chart
fig = px.line(df_trends, x='Month', y='Percent_Compliant',
              title='12-Month Compliance Maturity Trend',
              markers=True,
              text='Percent_Compliant',
              labels={'Percent_Compliant': 'Compliance %'})

# 3. Customize for a "Premium" look
fig.update_traces(line_color='#2E86C1', line_width=4, textposition="top center")
fig.update_layout(template='plotly_white', yaxis_range=[60, 100])

fig.show()



# Save the interactive chart as an HTML file
fig.write_html("compliance_trend.html")
```
